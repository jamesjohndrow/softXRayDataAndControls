## Initializing Data

### Loading Data

```python
dsource = psana.MPIDataSource('exp=sxrlq8415:run=192:smd')
acq_det=psana.Detector("Acq01")
mcp_traces = []
apd_traces = []
photon_energy = []
stop_event = 5000000
smldata = dsource.small_data('sxrlq8415run192_b.h5',gather_interval=100)
my_dict = {}
```

```python
for nevt,evt in enumerate(dsource.events()):
    if nevt>stop_event: break
    if (nevt%100==1): print nevt

    if(acq_det(evt) is None): continue

    mcp_wave = acq_det(evt)[0][2][1000:2000]-mean(acq_det(evt)[0][2][:300])
    apd_wave = acq_det(evt)[0][1][1000:2000]-mean(acq_det(evt)[0][1][:300])

    #my_dict['mcp_data'] = sum((abs(fft(mcp_wave[400:]))**1)[20:50]) #multiple ways of calculating area. which one is correct?
    my_dict['mcp_data'] = sum((abs(fft(mcp_wave[400:]))**2)[20:50])**0.5
    my_dict['apd_data'] = sum(apd_wave[190:230])
    
    smldata.event(my_dict)
    
smldata.save()
smldata.close()
```
